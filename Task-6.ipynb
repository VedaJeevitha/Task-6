# Prediction using Decision Tree Algorithm

**Importing Libraries**

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv(r'C:\Users\VEDA\Downloads\Iris.csv')
df.head()
```

```python
df.info()
```

```python
#removing the column Id 
df=df.drop(['Id'],axis=1)
```

```python
df.describe()
```

```python
#assigning values
x=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
```

```python
from sklearn.preprocessing import LabelEncoder
lab=LabelEncoder()
y=lab.fit_transform(y)
y
```

**Splitting training and testing data**

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y)
```

**Building Model**

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt=dt.fit(x_train,y_train)
print('Decision Tree Classifier Created')
```

```python
from sklearn import tree
fn=['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']
cn=['setosa','versicolor','virginica']

fig,axes=plt.subplots(nrows=1,ncols=1,figsize=(4,4),dpi=300)

tree.plot_tree(dt,feature_names=fn,class_names=cn,filled=True);
```

**Making Predictions**

```python
pred=dt.predict(x_test)
pred
```

**Comparing Actual vs Predicted Values**

```python
data=pd.DataFrame({'Actual':y_test,'Predicted':pred})
data.head()
```

**Accuracy Check**

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,pred)
```

__94% accuracy for this model__
